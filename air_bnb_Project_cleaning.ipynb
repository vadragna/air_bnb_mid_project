```python
columns_from_float_to_int = ['reviewer_id', 'review_id', 'bedrooms', 'square_feet', 'overall_rating', 'accuracy_rating', 'cleanliness_rating', 'checkin_rating', 'communication_rating', 'location_rating', 'value_rating']

def convert_from_float_into_int(x):
    return x.astype(int)

df[columns_from_float_to_int] = df[columns_from_float_to_int].apply(convert_from_float_into_int)

df.head()
```

Price is a string, should be a int

```python
import pandas as pd

def from_string_to_int(x):
    if ',' in x:
        x = x.replace(',', '') 
    value_before_period = x.split('.')[0]  
    float_value = float(x)  
    return int(float_value)

df['price'] = df['price'].apply(from_string_to_int)
```

review_date, host_since, first_review and last_review are strings, should be dates

In the comments column, I am replacing the encoded special characters of the most common European languages: German, French, Italian and Spanish.

```python
import pandas as pd
import re

replacements = {
    'Ã­': 'í',
    'Ã¡': 'á',
    'Ã±': 'ñ',
    'Ã©': 'é',
    'Ã¨': 'è',
    'Ã³': 'ó',
    'Ã¹ ': 'ù',
    'Ã': 'à',
    'Ã¼': 'ü',
    'Ã¤': 'ä',
    'Ã¶': 'ö',
    'A!': 'á',
    'A$?': 'ä',
    'A(c)': 'è',
    'A 1/4': 'ü',
    'A3': 'ó',
    'A"': 'è'
}

escaped_replacements = {re.escape(key): value for key, value in replacements.items()}

df['comments'] = df['comments'].replace(escaped_replacements, regex=True)
```

```python
df['reviewer_name'] = df['reviewer_name'].replace(escaped_replacements, regex=True)
```
